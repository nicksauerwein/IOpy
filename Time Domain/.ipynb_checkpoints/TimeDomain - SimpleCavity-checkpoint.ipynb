## Time Domain Simulation of a Simple Cavity


```python
omega_c = 5e9*np.pi*2
a = Mode('a', omega_c)
```

```python

kappa_ex = 0.2e6*np.pi*2
kappa_0 = 0.3e6*np.pi*2
kappa = kappa_ex + kappa_0


Delta =0
s_x = 1
s_y = 0

M = np.array([[-kappa/2, Delta],[-Delta, -kappa/2]])
L_ex = np.array([[np.sqrt(kappa_ex), 0],[0, np.sqrt(kappa_ex)]])

omega_ps = np.linspace(-10*kappa_0,10*kappa_0, 500)
Q = np.zeros_like(omega_ps)
P = np.zeros_like(omega_ps)

for i, omega_p in enumerate(omega_ps):
    def model(a, t):
        a_in = np.array([s_x * np.cos(omega_p*t), s_y * np.sin(omega_p*t)])
        dadt = M @ a + L_ex @ a_in
        return dadt
    a_0 = np.array([Q[i-1], P[i-1]])
    t = np.linspace(0, 5000 /30/kappa_0, 5000)
    a_in = np.array([np.cos(omega_p*t), np.sin(omega_p*t)])
    a = odeint(model, a_0, t)
    Q[i] = max(a_in.T[:,0] - np.sqrt(kappa_ex) * a[:,0])
    P[i] = max(a_in.T[:,1] - np.sqrt(kappa_ex) * a[:,1])
```

```python
plt.figure()
plt.plot(omega_ps/2/np.pi, Q/max(Q), label = '$X_{out}(\omega)$')
plt.xlabel('$\omega$ [Hz]')
plt.legend()
```
